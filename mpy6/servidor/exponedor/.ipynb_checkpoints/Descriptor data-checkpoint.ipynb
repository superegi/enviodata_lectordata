```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import interactive
import sqlite3
import time

%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Create your connection.
cnx = sqlite3.connect('../greenhouse.db')
df = pd.read_sql_query("SELECT * FROM readings", cnx)
```

```python
df.info()
df.head()
```

```python
df.Fecha = pd.to_datetime(df.Fecha)
df.Origen = df.Origen.str.replace('84:0d:8e:b0:f4:5c', 'TipoA')
df.Origen = df.Origen.str.replace('dc:4f:22:5f:4b:27', 'TipoB')
df.Origen = df.Origen.str.replace('dc:4f:22:5f:0d:2d', 'TipoC')

BD = df[['Origen', 'Fecha','Temperatura', 'Humedad']]
```

```python
BD.tail()
```

```python
BD.groupby('Origen').Temperatura.plot()
```

```python
BD.groupby('Origen').Humedad.plot()
```

solo un pequeno periodo

```python
# cantidad de horas para atrs
x = 20
dumm = BD.loc[
    (BD.Fecha > (pd.to_datetime(datetime.now()) - pd.Timedelta(hours=x)))
]
```

```python
dumm.groupby('Origen').Humedad.describe()
dumm.groupby('Origen').Humedad.plot(legend=True)
plt.title('Humedad')
```

```python
dumm.groupby('Origen').Temperatura.describe()
dumm.groupby('Origen').Temperatura.plot(legend=True)
plt.title('Temperatura')
```

```python
BD.groupby('Origen').plot(x='Fecha')
```
